```python
# 2. Preprocesamiento de los datos
# Verificamos si hay valores nulos
print("\n¿Hay valores nulos?", data.isnull().values.any())
```

```python
# 3. Selección de características
from sklearn.feature_selection import SelectKBest, f_classif
X = data.drop("quality", axis=1)
y = data["quality"]

# Seleccionar las 8 mejores características
selector = SelectKBest(score_func=f_classif, k=8)
X_new = selector.fit_transform(X, y)

# Ver qué columnas fueron seleccionadas
selected_features = X.columns[selector.get_support()]
print("\nCaracterísticas seleccionadas:")
print(selected_features)

# Usamos solo las columnas seleccionadas
X = data[selected_features]
```

```python
# 4. División Train/Test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# 5. Entrenamiento del modelo (Árbol de Decisión)
from sklearn.tree import DecisionTreeClassifier

# Crear modelo y entrenar
modelo = DecisionTreeClassifier(
    criterion='entropy', max_depth=5, min_samples_split=10, random_state=42
)
modelo.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Predecimos con el modelo
y_pred = modelo.predict(X_test)

# Mostramos métricas de evaluación
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nMatriz de Confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred))
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred, zero_division=0))
```

```python
# 7. Visualización de resultados
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(modelo, feature_names=selected_features, class_names=True, filled=True)
plt.title("Árbol de Decisión Entrenado")
plt.show()

# Importancia de características
importancia = pd.Series(modelo.feature_importances_, index=selected_features)
importancia.sort_values().plot(kind='barh', title='Importancia de Características')
plt.show()
```

```python
# 8. Interpretación y conclusión
print("Conclusiones:")
print("- El modelo de Árbol de Decisión fue entrenado con éxito.")
print("- Las características más relevantes incluyen:", list(selected_features))
print("- El rendimiento puede mejorarse ajustando hiperparámetros o probando otros modelos.")
```
